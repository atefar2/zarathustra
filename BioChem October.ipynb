```python
#3.11 MutableSeq objects
my_seq = Seq("GCCATTGTAATGGGCCGCTGAAAGGGTGCCCGA")

my_seq[5] = "G"
```

```python
mutable_seq = my_seq.tomutable()
mutable_seq

mutable_seq[5] = "C"
print(mutable_seq)
mutable_seq.remove("T")
print(mutable_seq)
mutable_seq.reverse()
print(mutable_seq)
```

```python
new_seq = mutable_seq.toseq()
new_seq
#a MutableSeq object just like from a Seq
```
